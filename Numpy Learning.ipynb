```python
# numpy.append.  add columns and rows to an exist array. axis = 0 for row, axis = 1 for column
import numpy as np
#show how to append a matrix to another
a = np.array([[1,2,3],[4,5,6]])
print 'original a is '
print a
b = np.append(a,[[7,9],[9,0]],axis=1)
print 'after appended a matrix, a has added two column to become b:'
print b
#let us try set axis = 0
c = np.append(a,[[7,9,9]],axis=0)
print 'after append another row in a, here is c'
print c
```

```python
# numpy.delete. remove colums or rows of an exist array. axis = 0 for row, axis = 1 for column
import numpy as np
a = np.array([[1,2,3,4],[5,6,7,8]])
print '\n original a is '
print a
b = np.delete(a,1,0) # 1 is for the 1st row, the count starts from 0
print '\n delete 1st row in a: '
print b

c = np.delete(a,np.s_[2:3:1],1) # remove the last column in a
print '\n delete last column in a:'
print c
```

```python
# as i cannot find ways to replace a string in numpy array. 
#i'm trying to preprocess the txt file( replace the string i want before i read it to numpy array)
import os
import datetime
# first read cluster file to get the map of the hashed address. address hash ----- address id
cluster_file = '/Users/ford-pro2/Machine-Learning/Didi/season_1/training_data/cluster_map/cluster_map'
fp = open(cluster_file,'r')
dict={}
for i in fp.readlines():
    result = i.split()
    dict[result[0]]=result[1]



def transDriverID(c):
    if c == 'NULL':
        return '0'
    else: 
        return '1'
def transDistID(c):
    return dict[c]
def transDate(c):
    year,month,day = c.split('-')
    #print datetime.datetime(int(year), int(month), int(day)).strftime('%w')
    return c.replace('-',' ')+' ' +datetime.datetime(int(year), int(month), int(day)).strftime('%w')
def transTime(c):
    hour,minutes,seconds=c.split(':')
    return str(int(hour)*6+int(minutes)/10+1)
def reassembleLine(line):
    # original line format: 'order_hash','driver_hash','passenger_hash','start_district_hash','dest_district_hash','price','date','time'
    # new line format: order_take start_district_id dest_district_id year month day week timeslot
    new_line = ''
    result = line.split()
    new_line+=transDriverID(result[1])+' '
    new_line+=transDistID(result[3])+' '
    new_line+=transDistID(result[4])+' '
    new_line+=transDate(result[6])+' '
    new_line+=transTime(result[7])+' '
    return new_line
    
order_file = '/Users/ford-pro2/Machine-Learning/Didi/order_sample'
order_translate = '/Users/ford-pro2/Machine-Learning/Didi/order_sample_translate'

fp = open(order_file,'r')  

wp = open(order_translate,'w')

for line in fp.readlines():
    transline =  reassembleLine(line)
    wp.writelines(transline+'\n')

wp.flush()
wp.close()
fp.close()

#for line in fp.readlines():
#    contents = line.split()
#    new_line += contents
    
```
